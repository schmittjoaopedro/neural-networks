```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

import sbs
```

```python
# Load Wine dataset
df_wine = pd.read_csv(
    'https://archive.ics.uci.edu/ml/'
    'machine-learning-databases/wine/wine.data',
    header=None
)

df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
                   'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']

# Classes are types of grapes grown in the same region from Italy
print('Class labels', np.unique(df_wine['Class label']))

df_wine.head()
```

```python
# Splitting the dataset into training and test sets
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values
X_train, X_test, y_train, y_test = \
    train_test_split(X, y,
                     test_size=0.3,  # 30% for test 
                     random_state=0,
                     # Stratify makes sure both train and test sets have the 
                     # same proportion of samples from different classes
                     stratify=y)
```

```python
# Normalization, also known as Min-Max scaling. In Min-Max scaling all values will be in the range [0, 1].
mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)

# Though min-max scaling is common, standardization is more robust as it centers de data around zero.
# For some optimization techniques like SGD, which initialize the weights around zero, standardization
# tends to make the learning of the parameters faster. Standardization does not change the shape of the
# distribution of the data.
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)

# Important to highlight that normalization and standardization are learned on train and applied in test
# and for every new data point without changing the initial parameters (re-learning).
```

```python
# L1 regularization leads to sparsity of the weights associated to each feature. This is given by its loss
# function diamond shaped which favors specific axes over the others. Because L1 favor some axes it ends up
# selecting a small subset of features which are relevant for the model, hence leading to a dimensionality
# reduction. In scikit-learn, L1 can be applied as follows:
lr = LogisticRegression(penalty='l1',  # L1-regularization
                        solver='liblinear',  # Optimization algorithm which supports L1
                        C=1.0,  # Default, decreasing it makes regularization stronger
                        multi_class='ovr')

lr.fit(X_train_std, y_train)  # Using standardized data
print('Training accuracy:', lr.score(X_train_std, y_train))
print('Test accuracy:', lr.score(X_test_std, y_test))

# Because we applied the model over a multiclass dataset, the first param/array of each output belongs to the
# respective class 1 versus 2 and 3, class 2 versus 1 and 3, and class 3 versus 1 and 2.
print("Bias params:", lr.intercept_)
# Each array corresponding to one class has 13 coefficients, one for each feature. Notice some features
# are set to zero, meaning they are not relevant for the model.
print("Weight params:", lr.coef_)
```

```python
# Printing the regularization path
fig = plt.figure()
ax = plt.subplot(111)
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black',
          'pink', 'lightgreen', 'lightblue', 'gray', 'indigo', 'orange']
weights, params = [], []
for c in np.arange(-4., 6.):
    lr = LogisticRegression(penalty='l1', C=10. ** c, solver='liblinear', multi_class='ovr', random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10 ** c)

weights = np.array(weights)
for column, color in zip(range(weights.shape[1]), colors):
    plt.plot(params, weights[:, column], label=df_wine.columns[column + 1], color=color)
plt.axhline(0, color='black', linestyle='--', linewidth=3)
plt.xlim([10 ** (-5), 10 ** 5])
plt.ylabel('Weight coefficient')
plt.xlabel('C (inverse regularization strength)')
plt.xscale('log')
plt.legend(loc='upper left')
ax.legend(loc='upper center', bbox_to_anchor=(1.38, 1.03), ncol=1, fancybox=True)
plt.show()
```

```python
# Testing the SBS algorithm (Sequential Backward Selection) to reduce the dimensionality of the dataset
knn = KNeighborsClassifier(n_neighbors=5)
sbs = sbs.SBS(knn, k_features=1)
sbs.fit(X_train_std, y_train)

k_feat = [len(k) for k in sbs.subsets_]
plt.plot(k_feat, sbs.scores_, marker='o')
plt.ylim([0.7, 1.02])
plt.ylabel('Accuracy')
plt.xlabel('Number of features')
plt.grid()
plt.tight_layout()
plt.show()
# Notice in the graph that the accuracy improves as the number of features decreases. This is because the
# course of dimensionality problem on KNNs algorithm, when fewer features are used more generalized the model
# becomes for the test set.
```

```python
# Let's see the 3-features with 100% accuracy
k3 = list(sbs.subsets_[10])
print(df_wine.columns[1:][k3])

# Performance of KNN on the original dataset
knn.fit(X_train_std, y_train)
print('Original Training accuracy:', knn.score(X_train_std, y_train))
print('Original Test accuracy:', knn.score(X_test_std, y_test))

# Performance of KNN on the reduced dataset
knn.fit(X_train_std[:, k3], y_train)
print('Reduced Training accuracy:', knn.score(X_train_std[:, k3], y_train))
print('Reduced Test accuracy:', knn.score(X_test_std[:, k3], y_test))

# Using less than a quarter of the features the model performance decreased slightly.
```

```python
# Random forests algorithms can also be used for feature selection
feat_labels = df_wine.columns[1:]
forest = RandomForestClassifier(n_estimators=500, random_state=1)
forest.fit(X_train, y_train)  # No need to use standardized or normalized data

importances = forest.feature_importances_  # Importance of each feature automatically calculated by the model
indices = np.argsort(importances)[::-1]  # Sort the features by importance
# Notice features important is normalized to their sum results 1
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))

plt.title('Feature importance')
plt.bar(range(X_train.shape[1]), importances[indices], align='center')
plt.xticks(range(X_train.shape[1]), feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

```python
forest = RandomForestClassifier(n_estimators=500, random_state=1)
forest.fit(X_train, y_train)

# This code uses Random Forest as a feature classifier intermediary step to 
# select the most important feature before passing it on to the main classifier
sfm = SelectFromModel(forest, threshold=0.1, prefit=True)
X_selected = sfm.transform(X_train)
print('Number of features that meet this threshold criterion:', X_selected.shape[1])
for f in range(X_selected.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
```
